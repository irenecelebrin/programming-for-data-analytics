```python
# import required packages 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv data from url using pandas: https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html 
url = 'https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download' 
data = pd.read_csv(url)

# display the first 5 rows of the dataframe
data.head(5)
```

```python
# uncomment to see information on the data
data.info()
```

```python
# create a new column with email domains (the part after the @ in the email address)
# select a subset of dataframe: https://pandas.pydata.org/docs/getting_started/intro_tutorials/03_subset_data.html 
# replace data in a df: https://pandas.pydata.org/docs/dev/reference/api/pandas.DataFrame.replace.html
data['Domains'] = data['Email'].replace(to_replace='^.+@', value='', regex=True)

data.head(5)
```

```python
# count occurances of each domain
domain_counts = data['Domains'].value_counts()

# show
domain_counts
```

```python
# extract arrays for domains and occurrrences 
# https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html, 
unique_domains = domain_counts.index     
n_domains = domain_counts.values
```

```python
# create the pie chart
# See: https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html
# and https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html

# set colors 
colors = ["paleturquoise", "lightcoral", "lemonchiffon"]

# plot the data
fig, ax = plt.subplots()
ax.pie(n_domains, labels = unique_domains, autopct='%1.1f%%', colors = colors)
plt.title('Email domains', loc='center', size= 16)

# show chart
plt.show()
```
