```python
# import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get the data from https://data.cso.ie/ API 
# see: https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html
df = pd.read_csv('https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FY006A/CSV/1.0/en')

# show first 5 rows
df.head(5)
```

```python
# Remove irrelevant columsn and rows 
# Select subset 
# Create pivot tables for every county 
# sum Male and Female population for every county, and calculate difference 
# Add all differences to an array 
# select max value fro the array 
```

```python
# List columns to remove irrelevant columns
headers = list(df)

headers
```

```python
# remove columns not needed for the analysis
col_to_drop = ['STATISTIC',
 'Statistic Label',
 'TLIST(A1)',
 'CensusYear',
 'C02199V02655',
 'C02076V03371',
 'C03789V04537',
 'UNIT']
df = df.drop(columns=col_to_drop)

df.head()
```

```python
# remove rows irrelevant rows (All sexes, All ages, Ireland)
df = df[df['Sex'] != 'Both sexes']
df = df[df['Single Year of Age'] != 'All ages']
df = df[df['Administrative Counties'] != 'Ireland']
# clean up Age values and make them integers 
df['Single Year of Age'] = df['Single Year of Age'].str.replace('Under 1 year', '0')
df['Single Year of Age'] = df['Single Year of Age'].str.replace('\D','', regex=True)
df['Single Year of Age'] = df['Single Year of Age'].astype(int)

# select subset where Single Year of Age is between 30 and 50 
age_target = 35
df = df[(df['Single Year of Age'] <= age_target + 5) & (df['Single Year of Age'] >= age_target - 5)]
df
```

```python
# Use subset to calculate age difference in each county (male - female)
# Create list for county name and list for difference
age_difference_by_county = []
county_name = []
unique_counties = df['Administrative Counties'].drop_duplicates()
for county in unique_counties:
    male_population = df.loc[(df['Administrative Counties']== county) & (df['Sex'] == 'Male')]['VALUE'].sum()
    female_population = df.loc[(df['Administrative Counties']== county) & (df['Sex'] == 'Female')]['VALUE'].sum()
    difference = male_population - female_population
    age_difference_by_county.append(difference)
    county_name.append(county)
```

```python
# Select highest age difference and corresponding county
max_diff = min(zip(county_name, age_difference_by_county), key=lambda x: x[1])
max_diff
```
